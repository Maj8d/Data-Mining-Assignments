```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv("cost_revenue_Cleaned.csv")
data
```

```python
duplicate = data.duplicated().sum()
duplicate
```

```python
nulls = data.isna()
nulls
```

```python
import matplotlib.pyplot as plt
```

```python
data.describe()
```

```python
x = pd.DataFrame(data, columns = ["Production_Budget"])
y = pd.DataFrame(data, columns = ['Worldwide_Gross'])

```

```python
plt.scatter(x,y, alpha = 0.3)
plt.xlabel("Production_Budget")
plt.ylabel("Worldwide_Gross")

plt.xlim(0,4.250000e+08)
plt.ylim(0,2.783919e+09)
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
linreg.fit(x,y)
```

```python
linreg.coef_
```

```python
linreg.intercept_
```

```python
plt.scatter(x,y, alpha = 0.3)
plt.xlabel("Production_Budget")
plt.ylabel("Worldwide_Gross")
plt.plot(x, linreg.predict(x), color='yellow')

plt.xlim(0, 4.250000e+08)
plt.ylim(0, 2.783919e+09)
```
