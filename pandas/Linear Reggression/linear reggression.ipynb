```python
x = pd.DataFrame(data, columns = ["Price"])
y = pd.DataFrame(data, columns = ['SquareFeet'])

```

```python
plt.scatter(x,y, alpha = 0.3)
plt.xlabel("Price")
plt.ylabel("SquareFeet")

plt.xlim(0, 492195.259972)
plt.ylim(0, 2999.000000)
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
linreg.fit(x,y)
```

```python
linreg.coef_
```

```python
linreg.intercept_
```

```python
plt.scatter(x,y, alpha = 0.3) 

plt.xlabel("Price")
plt.ylabel("SquareFeet")

plt.plot(x, linreg.predict(x), color='yellow')

plt.xlim(0, 492195.259972)
plt.ylim(0, 2999.000000)
```
